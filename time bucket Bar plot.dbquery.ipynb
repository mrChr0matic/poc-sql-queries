```sql
SELECT
    year,
    month,
    make_date(year, month, day) AS trip_date,
    time_bucket,

    CASE
        WHEN :metric_selector = 'revenue'
            THEN SUM(total_amount)

        WHEN :metric_selector = 'distance'
            THEN SUM(trip_distance)

        WHEN :metric_selector = 'passengers'
            THEN SUM(passenger_count)

        WHEN :metric_selector = 'trips'
            THEN COUNT(*)
    END AS selected_metric

FROM bi_taxi
GROUP BY year, month, trip_date, time_bucket
ORDER BY trip_date;
```

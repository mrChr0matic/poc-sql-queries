```sql
WITH base AS (
  SELECT
    year,
    month,
    make_date(year,month,day) AS trip_date,
    SUM(fare_amount) AS fare,
    SUM(tip_amount) AS tip,
    SUM(tolls_amount) AS tolls,
    SUM(extra) AS extra,
    SUM(mta_tax) AS mta_tax,
    SUM(improvement_surcharge) AS improvement
  FROM bi_taxi
  GROUP BY year,month,make_date(year,month,day)
)

SELECT year,month,trip_date,'Fare' AS fare_type,fare AS amount FROM base
UNION ALL
SELECT year,month,trip_date,'Tip',tip FROM base
UNION ALL
SELECT year,month,trip_date,'Tolls',tolls FROM base
UNION ALL
SELECT year,month,trip_date,'Extra',extra FROM base
UNION ALL
SELECT year,month,trip_date,'MTA Tax',mta_tax FROM base
UNION ALL
SELECT year,month,trip_date,'Improvement Surcharge',improvement FROM base;
```

```sql
WITH base AS (
  SELECT
    YEAR(tpep_pickup_datetime)  AS year,
    MONTH(tpep_pickup_datetime) AS month,
    DATE(tpep_pickup_datetime)  AS trip_date,

    ROUND(pickup_latitude,3)  AS pickup_lat,
    ROUND(pickup_longitude,3) AS pickup_lon,

    ROUND(dropoff_latitude,3)  AS dropoff_lat,
    ROUND(dropoff_longitude,3) AS dropoff_lon,

    COUNT(*) AS trip_count,
    SUM(total_amount) AS revenue

  FROM nyc_taxi_raw
  WHERE pickup_latitude BETWEEN 40.5 AND 41.0
    AND pickup_longitude BETWEEN -74.3 AND -73.6
    AND dropoff_latitude BETWEEN 40.5 AND 41.0
    AND dropoff_longitude BETWEEN -74.3 AND -73.6
  GROUP BY 1,2,3,4,5,6,7
)

SELECT *,
  CASE
    WHEN :kpi_selector='trips' THEN trip_count
    WHEN :kpi_selector='revenue' THEN revenue
  END AS selected_kpi
FROM base;
```

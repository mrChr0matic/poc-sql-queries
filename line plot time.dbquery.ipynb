```sql
SELECT *
FROM (

    SELECT
        make_date(year, month, day) AS trip_date,
        'revenue' AS metric_type,
        SUM(total_amount) AS metric_value
    FROM bi_taxi
    GROUP BY make_date(year, month, day)

    UNION ALL

    SELECT
        make_date(year, month, day),
        'distance',
        SUM(trip_distance)
    FROM bi_taxi
    GROUP BY make_date(year, month, day)

    UNION ALL

    SELECT
        make_date(year, month, day),
        'passengers',
        SUM(passenger_count)
    FROM bi_taxi
    GROUP BY make_date(year, month, day)

    UNION ALL

    SELECT
        make_date(year, month, day),
        'trips',
        COUNT(*)
    FROM bi_taxi
    GROUP BY make_date(year, month, day)

) t

WHERE metric_type = :metric_selector
ORDER BY trip_date;
```

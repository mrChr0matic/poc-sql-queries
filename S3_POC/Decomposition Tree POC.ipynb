```python
# dynamic drilling function
def get_manual_split(filters_dict, split_col, metric_col="Sales_Amount"):
    where_conditions = ["1=1"]
    for col, val in filters_dict.items():
        where_conditions.append(f"{col} = '{val}'")
    
    where_clause = " AND ".join(where_conditions)
    
    query = f"""
        SELECT 
            '{split_col}' as split_dimension, 
            {split_col} as node_name,        
            SUM({metric_col}) as value
        FROM poc_db.sales_gold
        WHERE {where_clause}
        GROUP BY {split_col}
        ORDER BY value DESC
    """
    
    print(f"--- Running Query ---\n{query}\n")
    
    return spark.sql(query)
```
